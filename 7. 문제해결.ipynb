## Up / Down 프로그램

```python
import random
print("======= Up / Down Program =======")
n = random.randint(1,100)
cnt=0

while True:
    cnt = cnt + 1
    print("Guess a number..")
    guess = int(input())
    if guess == n:
        print("Bingo !!")
        break
    elif guess < n:
        print("Up !!")
    else:
        print("Down !!")

print("You did it", cnt, "times")
```

## 반복 구조의 표현

```python
print("1에서 n까지의 합계")
n = int(input("n="))
sum = 0

for x in range(n+1):
    sum = sum + x;
    print(x, sum)
    
print("1에서 %d까지의 합계=%d" %(n, sum))
```

```python
print("a에서 b까지의 합계")
a = int(input("a="))
b = int(input("b="))
sum = 0

for x in range(a, b+1):
    sum = sum + x
    print(x, sum)
    
print("%d에서 %d까지의 합계=%d" %(a,b,sum))
```

```python
print("1에서 n까지의 홀수와 짝수의 합계")
n = int(input("n="))
sum1 = 0
sum2 = 0

for x in range(1, n+1):
    if (x % 2 == 1):
        sum1 = sum1 + x
    else:
        sum2 = sum2 + x
    
print("1에서 %d까지의 홀수의 합계=%d" %(n, sum1))
print("1에서 %d까지의 짝수의 합계=%d" %(n, sum2))
```

## 세 개의 숫자를 입력받아 가장 큰 수를 출력하기

```python
# 3개의 숫자를 입력받아 가장 큰 수를 출력하기
def max(a, b, c):    
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b 
    else:
        return c

a, b, c = input('3개의 숫자를 입력하세요: ').split()

n = max(a, b, c)

print('가장 큰 수는 : ', n)
```

## 세 개의 숫자를 입력받아 가장 작은 수를 출력하기

```python
# 3개의 숫자를 입력받아 가장 작은 수를 출력하기
def min(a, b, c):    
    if a<b and a<c:
        return a
    elif b<a and b<c:
        return b 
    else:
        return c

a, b, c = input('3개의 숫자를 입력하세요: ').split()

n = min(a, b, c)

print('가장 작은 수는 : ', n)

```

## 세 개의 숫자를 입력받아 중간 값 출력하기

```python
# 3개의 숫자를 입력받아 중간 값 출력하기
def mid(a, b, c):    
    if a>b>c or a<b<c:
        return b 
    elif a>c>b or a<c<b:
        return c
    elif b>a>c or b<a<c:
        return a

n = mid(10, 20, 30)
print('중간값은 : ', n)

n = mid(10, 30, 20)
print('중간값은 : ', n)

n = mid(20, 30, 10)
print('중간값은 : ', n)

n = mid(20, 10, 30)
print('중간값은 : ', n)

n = mid(30, 10, 20)
print('중간값은 : ', n)

n = mid(30, 20, 10)
print('중간값은 : ', n)
```

## 소수 판별하기

```python
# 소수 판별하기

n = 13
isPrime = True

for i in range(2, n):
    print(n, i, n%i)
    if (n % i == 0):    # n을 i로 나눈 나머지가 0이면
        isPrime = False
        break

if (isPrime == True):
    print(n, '은 소수입니다.')
else:
    print(n, '은 소수가 아닙니다.')

```

```python
# 소수 판별하기 - 함수

n = 49

def isPrime(n):
    for i in range(2, n):
        print(n, i, n%i)
        if (n % i == 0):
            return False
    return True

if (isPrime(n) == True):
    print(n, '은 소수입니다.')
else:
    print(n, '은 소수가 아닙니다.')
```

## 재귀함수 - 팩토리얼 출력하기

```python
# 팩토리얼 출력하기
def fact(n):
    if n == 1:
        return 1
    else:
        # fact(n) = n * fact(n-1)
        return n * fact(n-1)

print( fact(5) )
```

## 재귀함수 - 피보나치 수열 출력하기

```python
# 피보나치 수열 출력하기
def fib(n):
    if n == 0:      # 첫번째 항 : 0
        return 0
    elif n == 1:
        return 1    # 두번째 항 : 1
    else:
        # fib(n) = fib(n-1) + fib(n-2)
        return fib(n-1) + fib(n-2)
    
for i in range(10):
    print( fib(i), end=" " )
```

## 정수 뒤집어 출력하기

```python
n = int(input("정수를 입력하세요: "))
res=0
while n>0:
    r = n%10           # n을 10으로 나눈 나머지
    n = n//10          # n을 10으로 나눈 몫
    res = res*10 + r   # 자릿수를 올리고 나머지를 더함
    print(n, r, res)
print("결과는: ", res)
```
